# Task 1: Dataset Loading & Merging

```r
data_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
test_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test"

col_names <- c("age", "workclass", "fnlwgt", "education", "education_num",
               "marital_status", "occupation", "relationship", "race", "sex",
               "capital_gain", "capital_loss", "hours_per_week", "native_country",
               "income")

adult_data <- read.csv(data_url, header = FALSE, col.names = col_names, stringsAsFactors = FALSE)
adult_test <- read.csv(test_url, header = FALSE, col.names = col_names, stringsAsFactors = FALSE, skip = 1)

adult <- rbind(adult_data, adult_test)
```

# Task 6: Handling Missing Values for Workclass

```r
is_missing <- is.na(adult$workclass)
missing_values <- adult$workclass[is_missing]
head(missing_values, 10)

freq <- table(adult$workclass)
print(freq)

max_freq_index <- which.max(freq)
print(max_freq_index)

max_freq_name <- names(freq)[max_freq_index]
print(max_freq_name)

adult$workclass[is_missing] <- max_freq_name
sum(is.na(adult$workclass))
```

# Task 11: Outlier Detection and Capping for the Age Column

```r
summary(adult$age)

q1_age <- quantile(adult$age, 0.25)
q3_age <- quantile(adult$age, 0.75)
iqr_age <- q3_age - q1_age

lower_age <- q1_age - 1.5 * iqr_age
upper_age <- q3_age + 1.5 * iqr_age

is_age_outlier <- adult$age < lower_age | adult$age > upper_age
age_outliers_df <- adult[is_age_outlier, ]
nrow(age_outliers_df)
summary(age_outliers_df$age)

adult$age[adult$age > upper_age] <- upper_age

summary(adult$age)
```

# Task 12: Outlier Detection and Capping for Capital Gain Column

```r
summary(adult$capital_gain)
range(adult$capital_gain)

sum(adult$capital_gain == 99999)
summary(adult$capital_gain[adult$capital_gain < 99999])

next_highest_gain <- max(adult$capital_gain[adult$capital_gain < 99999])

adult$capital_gain[adult$capital_gain == 99999] <- next_highest_gain

summary(adult$capital_gain)
```

# Task 13: Outlier Detection and Capping for Capital Loss Column

```r
summary(adult$capital_loss)
range(adult$capital_loss)

sum(adult$capital_loss == 0)

summary(adult$capital_loss[adult$capital_loss > 0])

loss_nz <- adult$capital_loss[adult$capital_loss > 0]

q1_nz <- quantile(loss_nz, 0.25)
q3_nz <- quantile(loss_nz, 0.75)
iqr_nz <- q3_nz - q1_nz

print(iqr_nz)

upper_loss <- q3_nz + 1.5 * iqr_nz

sum(adult$capital_loss > upper_loss)
adult$capital_loss[adult$capital_loss > upper_loss] <- upper_loss

summary(adult$capital_loss)
```

# Task 14: Outlier Detection and Capping for Hours per Week

```r
summary(adult$hours_per_week)
range(adult$hours_per_week)

q1_hours <- quantile(adult$hours_per_week, 0.25)
q3_hours <- quantile(adult$hours_per_week, 0.75)
iqr_hours <- q3_hours - q1_hours

lower_hours <- q1_hours - 1.5 * iqr_hours
upper_hours <- q3_hours + 1.5 * iqr_hours

print(lower_hours)
print(upper_hours)

is_hours_outlier <- adult$hours_per_week < lower_hours | adult$hours_per_week > upper_hours
sum(is_hours_outlier)

hours_outliers_df <- adult[is_hours_outlier, ]
summary(hours_outliers_df$hours_per_week)

print(iqr_hours)
lower_hours <- q1_hours - 3 * iqr_hours
upper_hours <- q3_hours + 3 * iqr_hours

adult$hours_per_week[adult$hours_per_week < lower_hours] <- lower_hours
adult$hours_per_week[adult$hours_per_week > upper_hours] <- upper_hours

summary(adult$hours_per_week)
```

# Task 15: Data Transformation (Min-Max Normalization)

```r
min_max_norm <- function(x) {
  return ((x - min(x)) / (max(x) - min(x)))
}

adult$age <- min_max_norm(adult$age)
adult$capital_gain <- min_max_norm(adult$capital_gain)
adult$capital_loss <- min_max_norm(adult$capital_loss)
adult$hours_per_week <- min_max_norm(adult$hours_per_week)

summary(adult[c("age", "capital_gain", "capital_loss", "hours_per_week")])
```
