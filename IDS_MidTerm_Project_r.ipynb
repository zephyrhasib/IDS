# Task 1: Dataset Loading & Merging

```r
data_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
test_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test"

col_names <- c("age", "workclass", "fnlwgt", "education", "education_num",
               "marital_status", "occupation", "relationship", "race", "sex",
               "capital_gain", "capital_loss", "hours_per_week", "native_country",
               "income")

adult_data <- read.csv(data_url, header = FALSE, col.names = col_names, stringsAsFactors = FALSE)
adult_test <- read.csv(test_url, header = FALSE, col.names = col_names, stringsAsFactors = FALSE, skip = 1)

adult <- rbind(adult_data, adult_test)
```

# Task 5: Identification and Removal of Duplicate Records

```r
sum(duplicated(adult))

duplicates_df <- adult[duplicated(adult) | duplicated(adult, fromLast = TRUE), ]

duplicates_ordered <- duplicates_df[order(duplicates_df$fnlwgt, duplicates_df$age), ]

head(duplicates_ordered)

adult <- unique(adult)

sum(duplicated(adult))
```

# Task 6: Handling Missing Values for Workclass

```r
is_missing <- is.na(adult$workclass)
missing_values <- adult$workclass[is_missing]
head(missing_values, 10)

freq <- table(adult$workclass)
print(freq)

max_freq_index <- which.max(freq)
print(max_freq_index)

max_freq_name <- names(freq)[max_freq_index]
print(max_freq_name)

adult$workclass[is_missing] <- max_freq_name
sum(is.na(adult$workclass))
```

# Task 7: Handling Missing Values for Occupation

```r
is_missing_occ <- is.na(adult$occupation)
missing_values_occ <- adult$occupation[is_missing_occ]
head(missing_values_occ, 10)

freq_occ <- table(adult$occupation)
print(freq_occ)

max_freq_index_occ <- which.max(freq_occ)
print(max_freq_index_occ)

max_freq_name_occ <- names(freq_occ)[max_freq_index_occ]
print(max_freq_name_occ)

adult$occupation[is_missing_occ] <- max_freq_name_occ

sum(is.na(adult$occupation))
```

# Task 8: Handling Missing Values for Native Country

```r
is_missing_country <- is.na(adult$native_country)
missing_values_country <- adult$native_country[is_missing_country]
head(missing_values_country, 10)

freq_country <- table(adult$native_country)
print(head(sort(freq_country, decreasing = TRUE), 10))

max_freq_index_country <- which.max(freq_country)
print(max_freq_index_country)

max_freq_name_country <- names(freq_country)[max_freq_index_country]
print(max_freq_name_country)

adult$native_country[is_missing_country] <- max_freq_name_country

sum(is.na(adult$native_country))
```
